#### 實際練習 GPIOZero 控制 LED，確定單獨的 GPIO 控制 LED 亮跟暗交替閃爍完成，並且 PWM 控制 LED 明亮的完成後，嘗試依序改變led.value的值，分別設定 0.1, 0.3, 0.5, 0.7 觀察差異。

1.確定單獨的 GPIO 控制 LED 亮跟暗交替閃爍完成

```python
from gpiozero import PWMLED,LED
from signal import pause
from time import sleep
red=LED(17)
red.blink()
pause()
```

2.PWM 控制 LED 明亮的完成後，嘗試依序改變led.value的值，分別設定 0.1, 0.3, 0.5, 0.7 觀察差異

```python
from gpiozero import PWMLED,LED
from signal import pause
from time import sleep
red=PWMLED(17)
while True:
    red.value=0.1
    sleep(1)
    red.value=0.3
    sleep(1)
    red.value=0.5
    sleep(1)
    red.value=0.7
    sleep(1)
```

# 作業三

實際練習 GPIOZero 透過 Button 控制 LED，在按鈕的過程中，觀察實際按下按鈕的次數，LED 點亮的次數，是否一致。練習修改程式，讓按鈕按下是全亮，按鈕放開後是 30% 的亮度。

```python
from gpiozero import LED,Button,PWMLED
from signal import pause

led=PWMLED(17)
button=Button(2)
def led_1():
    led.value=1
def led_3():
    led.value=0.3
while True:
    button.when_pressed=led_1
    button.when_released=led_3

    pause()
```
